# Understanding the Important Concepts of Descriptive Statistics

Sample is a small part of information containing data of whole population
The following are the important topics which need to be discussed

1. **Mean**:- There are two means that are there sample mean and population mean.<br>Population mean is represented by $ \mu $ and sample bar is represented by $ x̄ $.<br>Mean is considered to be average of all the observations present in the data.<br>Formula for mean is (Sum of all observations)/(Total no. of observations)=$\frac{\sum_{i=1}^n(f_ix_i)}{(\sum_{i=1}^nf_i)}$ <br>where $f_i$ is the frequency of each observation$x_i$.<br>For class grouped data $x_i $ is taken as mid value of opening interval and closed interval, whereas in ungrouped data only exact numbers are taken.<br>


2. **Median**:- Median is the mid value of all observations taken one at a time(frequency of observations is not used).<br> For the ungrouped data mid term is taken as odd no. of observations and average of 2 mid terms in case of even no. of observations after arranging them in ascending and descending order. In case of grouped data the formula for fequency=<br>=$l+h\frac{\frac{n}{2}-c.f.}{f}$ where<br>$l$ = lower limit of median class<br>$n$ = number of observations<br>$c.f.$ = cumulative frequency of class preceding the median class<br>$f$ = frequency of median class<br>$h$ = class size


3. **Mode**:-It is the figure that has occurred maximum no. of counts. It can ve calculated both in grouped data and ungrouped data. For grouped data formula is<br>=$l+h\frac{f_1-f_0}{2f_1-f_0-f_2}$ where<br>$l$ = lower limit of the modal class<br>
$h$ = size of the class interval (assuming all class sizes to be equal)<br>
$f_1$ = frequency of the modal class<br>
$f_0$ = frequency of the class preceding the modal class<br>
$f_2$ = frequency of the class succeeding the modal class<br>

  **Relationship between Mean,Median and Mode<br>Mode=3Median−2Mean**
  
  
4. **Variance & Standard Deviation**:- It is a measure of the spread which determines how much observations vary from the mean. Calculating variance or standard variance is same for both grouped data and ungrouped data.<br>Formula of Variance=$\sigma^2=\frac{\sum_{i=1}^nf_i(x_i-\bar x)^2}{\sum_{i=1}^nf_i}$ <br>Formula of Standard Deviation=$\sigma=\sqrt\frac{\sum_{i=1}^nf_i(x_i-\bar x)^2}{\sum_{i=1}^nf_i}$<br>**Relation Between Variance and Standard Deviation<br>$Variance=(Standard Deviation)^2$<br>$Standard Deviation=\sqrt{Variance}$


5. **Covariance**:-It is a type of variance between two different random variables having(Assumption:-Having Same no. of observations) to the relation between them and represented by $\sigma(x,y)$.<br>Formula=$\sigma(x,y)$=**$\frac{\sum_{i=1}^nf_i(x_i-\bar x)(y_i-\bar y)}{\sum_{i=1}^nf_i}$**<br> where the random variables x_i and y_i are two different variables and f_i is the no. of observations.


6. **Correlation**:-It is the ratio of the covariance between two variables x_i and y_i and the product of standard deviation of two variables x_i and y_i and it is repersented bt $r$.<br>Formula=$\frac{Covariance(x,y)}{\sigma_x*\sigma_y}=\frac{\sum_{i=1}^nf_i(x_i-\bar x)(y_i-\bar y)}{\sum_{i=1}^nf_i(\sigma_x*\sigma_y)}$<br>The ratio of correlation is $-1 \leq r \leq 1$


7. **Skewness**:-It is distortion or asymmetry that deviates from the symmetrical bell curve, or normal distribution, in a set of data. If the curve is shifted to the left or to the right, it is said to be skewed and it is said to be skewed.It is represnted by $\beta_1$
<br>$\beta_1=\frac{\sum_{i=1}^n (f_i(x_i-\bar x))^3}{((\sum_{i=1}^nf_i)-1)*\sigma^3}=\frac{\mu_3^2}{\mu_2^3}$ where<br>$\mu_3$=Third Central Moment<br>$\mu_2$=Second Central Moment<br>There are three types of Skewness:-<br>1.Symmetric Skewness(Symmetrically Skewed) for $\beta_1=0$ which have **Mode=Median=Mean**<br>2.Positive Skewness(Positively Skewed) for $\beta_1>0$ which have **Mode<Median<Mean**<br>3.Negative Skewness(Negatively Skewed) for $\beta_1<0$ which have **Mode>Median>Mean**

8. **Kurtosis**:-It is the measure of the "tailedness" of the probability distribution of a real-valued random variable. It consists of Leptokurtic,Mesokurtic & Plastykurtic. It is represented by $\beta_2$<br>Formula=$\beta_2=\frac{\sum_{i=1}^n (f_i(x_i-\bar x))^4}{((\sum_{i=1}^nf_i)-1)*\sigma^4}=\frac{\mu_4}{\mu_2^4}$ where<br>$\mu_4=$Fourth Central Moment<br>$\mu_2=$Second Central Moment<br>There are three types of Kurtosis:-<br>1.Mesokurtic $(\beta_2=3)$-In this shape the curve is neither long tailed nor short tailed along y-axis.<br>2.Leptokurtic $(\beta_2>3)$-In this shape the curve is long tailed along y-axis.<br>3.Platykurtic $(\beta_2<3)$-In this shape the curve is short tailed along y-axis.

<img src="Types of Kurtosis.jpeg" width=500, height=1000>

9. **Normal Distribution**:-Normal distribution is a type of continuous probability distribution for a real-valued random variable. Normal distribution, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean. In graph form, normal distribution will appear as a bell curve.<br>$f(x)=\frac{1}{\sigma\sqrt{2\pi}}e^\frac{-(x-\mu)^2}{\sigma^2}$ where<br>$f(x)$=Probability Density Function<br>$x$=Random Variable(Sometimes in place of random variable Sample Mean $\bar x$<br>$\mu$=Population Mean(in case of Sample Mean $\bar x$ is taken)<br>$\sigma$=Standard Deviation<br>
Features<br>1.It is symmetric:-A normal distribution comes with a perfectly symmetrical shape. Distribution curve can be divided in the middle to produce two equal halves. The symmetric shape occurs when one-half of the observations fall on each side of the curve.<br>2.The mean, median, and mode are equal:-The middle point of a normal distribution is the point with the maximum frequency. The midpoint is also the point where these three measures fall. The measures are usually equal in a perfectly (normal) distribution.<br>3.Empirical Rule:-There is a constant proportion of distance lying under the curve between the mean and specific number of standard deviations from the mean.<br>4.Skewness and Kurtosis:-Skewness and kurtosis are coefficients that measure how different a distribution is from a normal distribution.<br>
Range of Normal Distribution=**[$\mu-3\sigma,\mu+3\sigma$]**<br>1.With 66.67% probabiity the amount can vary from mean is $\sigma$<br>2.With 95% probabiity the amount can vary from mean is $2\sigma$<br>3.With 99.99% probabiity the amount can vary from mean is $3\sigma$

10. **QQ Plot**:-A QQ plot is a probability plot, which is a graphical method for comparing two probability distributions by plotting their quantiles against each other. First, the set of intervals for the quantiles is chosen.<br>It follows tge particular process for it<br>1. Step 1:-Divide the data into equal no.of parts.<br>2. Step 2:-Divide the normal distribution curve into equal no. of parts(similar lines as divided in the Step 1.)<br>3. Step 3:-On graph having Y-Axis as data and X-Axis as normal distribution curve measures and mark the points on the graph with respective to data points and normal distribution measures.

11. **Box-Cox Transformation**:- It is the transformation in which one distribution is converted into another transformation.A Box Cox transformation is a transformation of a non-normal dependent variables into a normal shape. Normality is an important assumption for many statistical techniques; if your data isn't normal, applying a Box-Cox means that you are able to run a broader number of tests.
